## Creando nuevas columnas

```python
def golesvisita(marcador):
    goles = marcador.split('–')
    return int(goles[1])
def goleslocal(marcador):
    goles = marcador.split('–')
    return int(goles[0])

partidos6t['Goles Local'] = partidos6t['Score'].apply(goleslocal)
partidos6t['Goles Visitante'] = partidos6t['Score'].apply(golesvisita)
partidos6t['Total goles partido']= partidos6t['Goles Local']+partidos6t['Goles Visitante']

partidos6t
```

```python
def torneo(año,mes):
    if ((año==2016) & (mes<=12)|(año==2017) & (mes<8)):
        return '16/17'
    elif ((año==2017) & (mes<=12)|(año==2018) & (mes<8)):
        return '17/18'
    elif ((año==2018) & (mes<=12)|(año==2019) & (mes<8)):
        return '18/19'
    elif ((año==2019) & (mes<=12)|(año==2020) & (mes<9)):
        return '19/20'
    elif ((año==2020) & (mes<=12)|(año==2021) & (mes<8)):
        return '20/21'
    elif ((año==2021) & (mes<=12)|(año==2022) & (mes<8)):
        return '21/22'

    
partidos6t['Año'] = partidos6t['Date'].dt.year
partidos6t['Mes'] = partidos6t['Date'].dt.month
partidos6t['Temporada'] = partidos6t.apply(lambda x: torneo(x['Año'],x['Mes']),axis = 1)

partidos6t
```

```python
def resultado(string):
    goles = string.split('–')
    if int(goles[0])>int(goles[1]):
        return 'Local'
    if int(goles[0])<int(goles[1]):
        return 'Visitante'
    else: 
        return 'Empate'
    
def resultado_equipo(string, local, visita):
    goles = string.split('–')
    if int(goles[0])>int(goles[1]):
        return local
    if int(goles[0])<int(goles[1]):
        return visita
    else: 
        return 'Empate'
    
partidos6t['Resultado'] = partidos6t['Score'].apply(resultado)
partidos6t['Ganador'] = partidos6t.apply(lambda x: resultado_equipo(x['Score'],x['Local'],x['Visitante']),axis=1)

partidos6t

```

## Tabla de posiciones

```python
ej2122 = partidos6t[(partidos6t['Temporada']=='21/22')&(partidos6t['Torneo']=='Primera división')]
equip2122 = ej2122['Local'].unique().tolist()
equip2122
```

```python
def tabla_posicion(data,fecha,equipos):
    tabla = {}
    for equipo in equipos:
        data_usar = data[(data['Wk']<=fecha)&((data['Local']==equipo)|(data['Visitante']==equipo))]
        puntos = 0
        for i in range(len(data_usar)):
            ganador = str(data_usar.iloc[i]['Ganador'])
            if ganador == equipo:
                puntos += 3
            if ganador == 'Empate':
                puntos +=1
        tabla[equipo] = puntos
    df = pd.DataFrame([[key, tabla[key]] for key in tabla.keys()], columns=['Equipo', 'Puntos'])
    df = df.sort_values(by='Puntos',ascending = False)
    return df
```

```python
import pandas as pd

# Supongamos que tienes tu DataFrame llamado 'partidos6t'
# Primero, filtra los partidos de la temporada 21/22 de la primera división
partidos_21_22 = partidos6t[(partidos6t['Temporada'] == '21/22') & (partidos6t['Torneo'] == 'Primera división')]

# Inicializa las variables para llevar un conteo de los goles en contra
goles_concedidos_osasuna = 0
goles_concedidos_sevilla = 0

# Itera a través de cada partido
for index, partido in partidos_21_22.iterrows():
    # Verifica si Osasuna jugó como local y si la fecha (Wk) es hasta la fecha 24
    if partido['Local'] == 'Osasuna' and partido['Wk'] <= 24:
        goles_concedidos_osasuna += partido['Goles Visitante']
    # Verifica si Osasuna jugó como visitante y si la fecha (Wk) es hasta la fecha 24
    elif partido['Visitante'] == 'Osasuna' and partido['Wk'] <= 24:
        goles_concedidos_osasuna += partido['Goles Local']
    
    # Repite el proceso para el Sevilla
    if partido['Local'] == 'Sevilla' and partido['Wk'] <= 24:
        goles_concedidos_sevilla += partido['Goles Visitante']
    elif partido['Visitante'] == 'Sevilla' and partido['Wk'] <= 24:
        goles_concedidos_sevilla += partido['Goles Local']

# Imprime los resultados
print("Goles concedidos a Osasuna hasta la fecha 24: ", goles_concedidos_osasuna)
print("Goles concedidos al Sevilla hasta la fecha 24: ", goles_concedidos_sevilla)
```

## **Pregunta 4**
 
Considerando la temporada 21/22 de la primera división del fútbol español, ¿cuál de las siguientes 
alternativas indica la cantidad total de goles que ha convertido el Elche y la Real sociedad desde la fecha 1 a 
la fecha 20 (ambas incluidas)? 
 


```python
import pandas as pd

# Supongamos que tienes tu DataFrame llamado 'partidos6t'
# Primero, filtra los partidos de la temporada 21/22 de la primera división
partidos_21_22 = partidos6t[(partidos6t['Temporada'] == '21/22') & (partidos6t['Torneo'] == 'Primera división')]

# Inicializa las variables para llevar un conteo de los goles convertidos por Elche y Real Sociedad
goles_convertidos_elche = 0
goles_convertidos_real_sociedad = 0

# Itera a través de cada partido
for index, partido in partidos_21_22.iterrows():
    # Verifica si Elche jugó como local y si la fecha (Wk) es hasta la fecha 20
    if partido['Local'] == 'Elche' and partido['Wk'] <= 20:
        goles_convertidos_elche += partido['Goles Local']
    # Verifica si Elche jugó como visitante y si la fecha (Wk) es hasta la fecha 20
    elif partido['Visitante'] == 'Elche' and partido['Wk'] <= 20:
        goles_convertidos_elche += partido['Goles Visitante']
    
    # Repite el proceso para la Real Sociedad
    if partido['Local'] == 'Real Sociedad' and partido['Wk'] <= 20:
        goles_convertidos_real_sociedad += partido['Goles Local']
    elif partido['Visitante'] == 'Real Sociedad' and partido['Wk'] <= 20:
        goles_convertidos_real_sociedad += partido['Goles Visitante']

# Imprime los resultados
print("Goles convertidos por Elche hasta la fecha 20: ", goles_convertidos_elche)
print("Goles convertidos por Real Sociedad hasta la fecha 20: ", goles_convertidos_real_sociedad)
```
