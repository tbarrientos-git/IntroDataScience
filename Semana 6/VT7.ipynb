## Rellenar valores faltantes en Descuento

```python
boleta.loc[:,'Descuento'] = boleta.loc[:,'Descuento'].fillna('-')
```

```python
boleta_dsctofalt = boleta[(boleta['Descuento']=='-')]
```

```python
boleta_dsctofalt
```

```python
def precio(p1,p2,p3,p4,p5,p6,p7,p8,p9,p10,precios):
    productos = [p1,p2,p3,p4,p5,p6,p7,p8,p9,p10]
    precio = 0
    i=0
    for producto in productos:
        if producto == '-':
            pass
        else:
            data = precios[precios['Producto']==producto]
            precio += int(precios[precios['Producto']==producto]['Precio'])        
    return precio    
```

```python
boleta["Suma precio neto"] = boleta.apply(lambda x: precio(x['Producto 1'],x['Producto 2'],x['Producto 3'],
                            x['Producto 4'],x['Producto 5'],x['Producto 6'],x['Producto 7'],x['Producto 8']
                            ,x['Producto 9'],x['Producto 10'],prod),axis = 1)
```

```python
def descuento(dscto,v_bol,v_neto):
    if dscto =='-':
        return np.round(1-v_bol/v_neto,2)
    else:
        return dscto    
```

```python
boleta['Descuento']=boleta.apply(lambda x: descuento(x['Descuento'],x['Valor Boleta'],x['Suma precio neto']),axis=1)
```

## Rellenar valores faltantes en Valor Boleta

```python
boleta.isna().sum()
```

```python
boleta['Valor Boleta']= boleta['Valor Boleta'].fillna('*')
```

```python
boleta[boleta['Valor Boleta']=='*']
```

```python
def valor(dscto,v_bol,v_neto):
    if v_bol == '*':
        return np.round(v_neto*(1-dscto),2)
    else:
        return v_bol    
```

```python
boleta['Valor Boleta']= boleta.apply(lambda x: valor(x['Descuento'],x['Valor Boleta'],x['Suma precio neto']),axis=1)
```
